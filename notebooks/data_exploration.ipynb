```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pyedflib as edf
```

### Loading Data

```python
file = "rawdata/bidsexport/sub-001/ses-01/eeg/sub-001_ses-01_task-meditation_eeg.bdf"
```

```python
f = edf.EdfReader(file)
n = f.signals_in_file
signal_labels = f.getSignalLabels()
sigbufs = np.zeros((n, f.getNSamples()[0]))
for i in np.arange(n):
    sigbufs[i, :] = f.readSignal(i)
```

```python
signals = pd.DataFrame(data=sigbufs).T
signals.columns=signal_labels
```
